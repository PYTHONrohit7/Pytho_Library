```python
x=np.random.rand(50) 
y=np.random.rand(50)
plt.scatter(x,y)
```

```python
colours=np.random.rand(50)
size=1000*np.random.rand(50)
plt.scatter(x,y,c=colours,s=size,alpha=.2,marker="v")    #(x-axis,y-axis,colour,size,alpha visiblity,marker shape of dot)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Random data")
```

## Beautification of Graph

```python
x=np.linspace(0,10,200)
```

```python
y=np.sin(x)
```

```python
plt.xlabel("This is X axis")
plt.ylabel("This is y axis")
plt.title("The title of graph")
plt.plot(x,y,"--r")
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,"--b")
plt.xlabel("This is X axis")
plt.ylabel("This is y axis")
plt.title("The title of graph")
plt.show()
```
